## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
import matplotlib.pyplot as plt

# Visualizing data from the first table: aqsuitisions_df

num_aquirers = acquisitions_df['acquiring_object_id'].nunique()
# print(num_aquirers)

# Seems there's too many different aquirers to make a helpful visualization using them as an axis

# Plotting how many acquisitions are of each currency type. 
acquisitions_by_currency = acquisitions_df['price_currency_code'].value_counts()
print(acquisitions_by_currency)

plt.figure(figsize=(10, 6))
acquisitions_by_currency.plot(kind='bar')
plt.title('Aquisitions by Currency Type')
plt.xlabel('Currency Type')
plt.ylabel('Count')
plt.show()

# I thought it could be helpful to see how many aquisitions came from different parts of the world,
# as evidenced by the currency used.  However, such a proponderance are in USD that this doesn't 
# seem that helpful of a visualization, after all. 
```

```python
# Visualizing data from the second table: degrees_df

degrees_by_type = degrees_df['degree_type'].value_counts()
# print(degrees_by_type)

# That's too many to plot

top_10_degrees = degrees_by_type.sort_values(ascending=False).head(10)
print(top_10_degrees)

plt.figure(figsize=(10, 6))
top_10_degrees.plot(kind='bar')
plt.title('Aquisitions by Degree Type')
plt.xlabel('Degree Type')
plt.ylabel('Count')
plt.show()
```

```python
# Visualizing data from the third table: funding_rounds_df

# Visualization for how much funding was received by funding round type: 

rounds_grouped = funding_rounds_df.groupby('funding_round_type')['raised_amount_usd'].sum().reset_index()
# print(rounds_grouped.shape)
print(rounds_grouped.dtypes)
print(funding_rounds_df['funding_round_type'].unique())

plt.figure(figsize=(10, 6))
plt.bar(rounds_grouped['funding_round_type'], rounds_grouped['raised_amount_usd'])
plt.title('Funding Received by Funding Type')
plt.xlabel('Funding Type')
plt.ylabel('Funding Received (USD)')
plt.ticklabel_format(style='plain', axis='y')
plt.show()

# the y-axis isn't the most legiblae.  I think that setting the style to 'plain' is still better
# than it saying 1.2 x 1e11, for example, but commas would make this a lot more understandable. 
```

```python
# Visualizing data from the fourth table: funds_df

# Visualization for which companies make up the top 10 of most funding received
top_10_companies = funds_df.groupby('name')['raised_amount'].sum().nlargest(10).reset_index()
print(top_10_companies)

plt.figure(figsize=(10, 6))
plt.bar(top_10_companies['name'], top_10_companies['raised_amount'])
plt.title('Top 10 Companies by Funding Raised')
plt.xlabel('Company Name')
plt.ylabel('Amount Raised')
plt.xticks(rotation=90)
plt.ticklabel_format(style='plain', axis='y')
plt.show()

# Visualization for how many 
```

```python
# Visualizing data from the fifth table: investments_df

# Vizualition for number of deals over time

# Seeing if the 'created_at' column's data is in datetime format
print(investments_df['created_at'].dtype)

# It's not, so converting to datetime format
investments_df['created_at'] = pd.to_datetime(investments_df['created_at'])

investments_df['year'] = investments_df['created_at'].dt.year
print(investments_df['year'].head())

funding_by_year = investments_df.groupby('year').size().reset_index(name='num_deals')

plt.figure(figsize=(10, 6))
plt.plot(funding_by_year['year'], funding_by_year['num_deals'])
plt.title('Trend of Deals Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Deals')
plt.show()
```

```python
# Visualizing data for the sixth table: ipos_df

# Visualization for the amount raised for the top 10 / 25 ipos: 

# Checking to see if there are any NaNs in the 'raised_amount' column
raised_nans = ipos_df['raised_amount'].isna().sum()

# It would appear that there aren't any Nans
print(raised_nans)

# Want to limit this to the top 10 ipos
top_25_ipos = ipos_df.sort_values(by='raised_amount', ascending=False).head(10)
print(top_25_ipos)

# Visualization

plt.figure(figsize=(10, 6))
plt.bar(top_25_ipos['stock_symbol'], top_25_ipos['raised_amount'])
plt.title('Top 10 IPOs by Amount Raised')
plt.xlabel('IPO by Stock Symbol')
plt.ylabel('Amount Raised (USD)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualizing data for the seventh table: milestones_df

# Visualization for the number of milestones per year 

# Converting the 'created_at' column to useable datetime format
milestones_df['created_at'] = pd.to_datetime(milestones_df['created_at'])

# Extracting the years from the datetimes
milestones_df['year'] = milestones_df['created_at'].dt.year
print(milestones_df['year'].head())

# Grouping by years
milestones_by_year = milestones_df.groupby('year').size().reset_index(name='num_milestones')

# Plotting 
plt.figure(figsize=(10, 6))
plt.plot(milestones_by_year['year'], milestones_by_year['num_milestones'])
plt.title('Trend of Milestones Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Milestones')
plt.show()

```

```python
# Visualizing data for the eighth table: objects_df

categories = objects_df['category_code'].unique()
print(categories)

# Visualization for the total amount of funding for each category code

funding_by_category = objects_df.groupby('category_code')['funding_total_usd'].sum().reset_index()
print(funding_by_category)

plt.figure(figsize=(10, 6))
plt.bar(funding_by_category['category_code'], funding_by_category['funding_total_usd'])
plt.title('Funding Received by Category')
plt.xlabel('Category')
plt.ylabel('Funding Received')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualizing data for the ninth table: offices_df

# Visualization comparing the representation of different regions
num_per_region = offices_df.groupby('region').size().reset_index(name='num_offices')
print(num_per_region)

# That's too many to use... going to limit to top 10
top_10_regions = num_per_region.sort_values(by='num_offices', ascending=False).head(10)
print(top_10_regions)

plt.figure(figsize=(10, 6))
plt.bar(top_10_regions['region'], top_10_regions['num_offices'])
plt.title('Number of Offices by Region')
plt.xlabel('Regions')
plt.ylabel('Number of Offices')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualizing data for the tenth table: people_df

# Visualization of top 25 birthplaces
num_by_birthplace = people_df['birthplace'].value_counts().head(25)
print(num_by_birthplace)

plt.figure(figsize=(10, 6))
plt.bar(num_by_birthplace.index, num_by_birthplace.values)
plt.title('Top 25 Most Represented Birthplaces')
plt.xlabel('Birthplace')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualizing data for the eleventh table: relationships_df

# Visualizaton of the top 25 most common titles
num_by_title = relationships_df['title'].value_counts().head(25)
print(num_by_title)

plt.figure(figsize=(10, 6))
plt.bar(num_by_title.index, num_by_title.values)
plt.title('Most Common Roles')
plt.xlabel('Titles')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
